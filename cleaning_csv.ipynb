```python
def_stats[['Total Tackles', 'Solo Tackles', 'Assisted Tackles', 'Sacks', 'Safties',
       'Passes Defended', 'Ints', 'Ints for TDs', 'Int Yards', 'Yards Per Int',
       'Longest Int Return']] = def_stats[['Total Tackles', 'Solo Tackles', 'Assisted Tackles', 'Sacks', 'Safties',
       'Passes Defended', 'Ints', 'Ints for TDs', 'Int Yards', 'Yards Per Int',
       'Longest Int Return']].apply(pd.to_numeric, errors='coerce')
def_stats.dtypes
```

```python
def_stats_gb = def_stats.groupby(['Player Id'])
```

```python
defense = def_stats_gb[['Games Played',
       'Total Tackles', 'Solo Tackles', 'Assisted Tackles', 'Sacks', 'Safties',
       'Passes Defended', 'Ints', 'Ints for TDs', 'Int Yards', 'Yards Per Int',
       'Longest Int Return']].sum()
defense
```

```python
# Sum of these defensive stats  **must reset_index() to keep Player Id as a column
defense = def_stats_gb[['Games Played',
       'Total Tackles', 'Solo Tackles', 'Assisted Tackles', 'Sacks', 'Safties',
       'Passes Defended', 'Ints', 'Ints for TDs', 'Int Yards', 'Yards Per Int',
       'Longest Int Return']].sum().reset_index()

print(defense.head())

# Max of these defensive stats
defense1 = def_stats_gb[['Yards Per Int',
       'Longest Int Return']].max().reset_index()

print(defense1.head())

# Recombine into new stats DF
defense_df = pd.merge(defense, defense1, on='Player Id', how='inner')
defense_df
```
