```python
import pandas as pd
from sqlalchemy import create_engine
import sqlite3
from sqlite3 import Error
```

```python
basic_file = 'Resources/Basic_Stats.csv'
basic_stats = pd.read_csv(basic_file)
```

```python
def_file = 'Resources/Career_Stats_Defensive.csv'
def_stats = pd.read_csv(def_file)
```

```python
pass_file = 'Resources/Career_Stats_Passing.csv'
pass_stats = pd.read_csv(pass_file)
```

```python
rec_file = 'Resources/Career_Stats_Receiving.csv'
rec_stats = pd.read_csv(rec_file)
```

```python
rush_file = 'Resources/Career_Stats_Rushing.csv'
rush_stats = pd.read_csv(rush_file)
```

```python
basic_stats_df = basic_stats[['Player Id',
                                   'Name',
                                   'Age', 
                                   'Birth Place', 
                                   'Birthday', 
                                   'College', 
                                   'Height (inches)', 
                                   'Weight (lbs)']]
```

```python
id = basic_stats_df['Player Id'].str.split('/', n=1, expand=True)
basic_stats_df['id'] = id[1]

name = basic_stats_df['Name'].str.split(', ', n=1, expand=True)
basic_stats_df['First Name'] = name[1]
basic_stats_df['Last Name'] = name[0]
```

```python
basic_stats_df = basic_stats_df[['id',
               'First Name',
                'Last Name',
               'Age', 
               'Birth Place', 
               'Birthday', 
               'College', 
               'Height (inches)', 
               'Weight (lbs)']]
```

```python
basic_stats_df = basic_stats_df.rename(columns={'First Name':'First_Name',
                       'Last Name':'Last_Name',
                       'Birth Place':'Birth_Place', 
                        'Height (inches)': 'Height(inches)', 
                        'Weight (lbs)':'Weight(lbs)'})
basic_stats_df.head()
```

```python
id = def_stats['Player Id'].str.split('/', n=1, expand=True)
def_stats['id'] = id[1]

name = def_stats['Name'].str.split(', ', n=1, expand=True)
def_stats['First Name'] = name[1]
def_stats['Last Name'] = name[0]
```

```python
def_stats_df = def_stats[['id',
                          'First Name',
                          'Last Name',
                          'Position',
                          'Year',
                          'Team',
                          'Games Played',
                        'Total Tackles',
                          'Solo Tackles',
                          'Assisted Tackles',
                          'Sacks',
                          'Safties',
                           'Passes Defended',
                          'Ints',
                          'Ints for TDs',
                          'Int Yards',
                          'Yards Per Int',
                           'Longest Int Return']]
```

```python
def_stats_df = def_stats_df.rename(columns={'First Name':'First_Name',
                          'Last Name':'Last_Name',
                          'Games Played':'Games_Played',
                        'Total Tackles':'Total_Tackles',
                          'Solo Tackles':'Solo_Tackles',
                          'Assisted Tackles':'Assisted_Tackles',
                           'Passes Defended':'Passes_Defended',
                          'Ints for TDs':'Ints_for_TDs',
                          'Int Yards':'Int_Yards',
                          'Yard Per Int':'Yards_Per_Int',
                           'Longest Int Return':'Longest_Int_Return'})
def_stats_df.head()
```

```python
def_stats_df[[
           'Total_Tackles', 
              'Solo_Tackles', 
              'Assisted_Tackles',
               'Sacks', 
              'Safties', 
              'Passes_Defended', 
              'Ints', 
              'Ints_for_TDs',
               'Int_Yards', 
              'Longest_Int_Return']] = def_stats_df[[
           'Total_Tackles', 
              'Solo_Tackles', 
              'Assisted_Tackles',
               'Sacks', 
              'Safties', 
              'Passes_Defended', 
              'Ints', 
              'Ints_for_TDs',
               'Int_Yards', 
              'Longest_Int_Return']].apply(pd.to_numeric, errors='coerce')
```

```python
id = pass_stats['Player Id'].str.split('/', n=1, expand=True)
pass_stats['id'] = id[1]

name = pass_stats['Name'].str.split(', ', n=1, expand=True)
pass_stats['First Name'] = name[1]
pass_stats['Last Name'] = name[0]
```

```python
pass_stats_df = pass_stats[['id',
                            'First Name',
                            'Last Name',
                            'Position',
                            'Year', 'Team',
                            'Games Played',
                            'Passes Attempted',
                            'Passes Completed',
                            'Completion Percentage',
                           'Pass Attempts Per Game',
                            'Passing Yards',
                            'Passing Yards Per Attempt',
                           'Passing Yards Per Game',
                            'TD Passes',
                            'Percentage of TDs per Attempts',
                               'Ints',
                            'Int Rate',
                            'Longest Pass',
                            'Passes Longer than 20 Yards',
                               'Passes Longer than 40 Yards',
                            'Sacks',
                            'Sacked Yards Lost',
                            'Passer Rating']]
```

```python
pass_stats_df = pass_stats_df.rename(columns={'First Name':'First_Name',
                            'Last Name':'Last_Name',
                            'Games Played':'Games_Played',
                            'Passes Attempted':'Passes_Attempted',
                            'Passes Completed':'Passes_Completed',
                            'Completion Percentage':'Completion_Percentage',
                            'Pass Attempts Per Game':'Pass_Attempts_Per_Game',
                            'Passing Yards':'Passing_Yards',
                            'Passing Yards Per Attempt':'Passing_Yards_Per_Attempt',
                            'Passing Yards Per Game':'Passing_Yards_Per_Game',
                            'TD Passes':'TD_Passes',
                            'Percentage of TDs per Attempts':'Percentage_of_TDs_per_Attempts',
                            'Int Rate':'Int_Rate',
                            'Longest Pass':'Longest_Pass',
                            'Passes Longer than 20 Yards':'Passes_Longer_than_20_Yards',
                            'Passes Longer than 40 Yards':'Passes_Longer_than_40_Yards',
                            'Sacked Yards Lost':'Sacked_Yards_Lost',
                            'Passer Rating':'Passer_Rating'})
pass_stats_df.head()
```

```python
pass_stats_df[['Passes_Attempted', 'Passes_Completed',
       'Completion_Percentage', 'Passing_Yards',
       'Passing_Yards_Per_Attempt', 'Passing_Yards_Per_Game', 'TD_Passes',
       'Percentage_of_TDs_per_Attempts', 'Ints', 'Int_Rate', 'Longest_Pass',
       'Passes_Longer_than_20_Yards', 'Passes_Longer_than_40_Yards', 'Sacks',
       'Sacked_Yards_Lost']] = pass_stats_df[['Passes_Attempted', 'Passes_Completed',
       'Completion_Percentage', 'Passing_Yards',
       'Passing_Yards_Per_Attempt', 'Passing_Yards_Per_Game', 'TD_Passes',
       'Percentage_of_TDs_per_Attempts', 'Ints', 'Int_Rate', 'Longest_Pass',
       'Passes_Longer_than_20_Yards', 'Passes_Longer_than_40_Yards', 'Sacks',
       'Sacked_Yards_Lost']].apply(pd.to_numeric, errors='coerce')
```

```python
id = rush_stats['Player Id'].str.split('/', n=1, expand=True)
rush_stats['id'] = id[1]

name = rush_stats['Name'].str.split(', ', n=1, expand=True)
rush_stats['First Name'] = name[1]
rush_stats['Last Name'] = name[0]
```

```python
rush_stats_df = rush_stats[['id', 
                            'First Name', 
                            'Last Name',
                            'Position', 
                            'Year', 
                            'Team', 
                            'Games Played',
                            'Rushing Attempts', 
                            'Rushing Attempts Per Game', 
                            'Rushing Yards',
                           'Yards Per Carry', 
                            'Rushing Yards Per Game', 
                            'Rushing TDs',
                           'Longest Rushing Run', 
                            'Rushing First Downs',
                           'Percentage of Rushing First Downs', 
                            'Rushing More Than 20 Yards',
                            'Rushing More Than 40 Yards', 
                            'Fumbles']]
```

```python
rush_stats_df = rush_stats_df.rename(columns = {'First Name':'First_Name',
                    'Last Name':'Last_Name',
                    'Games Played':'Games_Played',
                    'Rushing Attempts':'Rushing_Attempts',
                    'Rushing Attempts Per Game':'Rushing_Attempts_Per_Game',
                    'Rushing Yards':'Rushing_Yards',
                    'Yards Per Carry':'Yards_Per_Carry',
                    'Rushing Yards Per Game':'Rushing_Yards_Per_Game',
                    'Rushing TDs':'Rushing_TDs',
                    'Longest Rushiing Run':'Longest_Rushing_Run',
                    'Rushing First Downs':'Rushing_First_Downs',
                    'Percentage of Rushing First Downs':'Percentage_of_Rushing_First_Downs',
                    'Rushing More Than 20 Yards':'Rushing_More_Than_20_Yards',
                    'Rushing More Than 40 Yards':'Rushing_More_Than_40_Yards'})
```

```python
rush_stats_df[['Rushing_Attempts',
       'Rushing_Yards', 'Yards_Per_Carry', 'Rushing_Yards_Per_Game',
       'Rushing_TDs', 'Longest Rushing Run', 'Rushing_First_Downs',
       'Percentage_of_Rushing_First_Downs', 'Rushing_More_Than_20_Yards',
       'Rushing_More_Than_40_Yards', 'Fumbles']] = rush_stats_df[['Rushing_Attempts',
       'Rushing_Yards', 'Yards_Per_Carry', 'Rushing_Yards_Per_Game',
       'Rushing_TDs', 'Longest Rushing Run', 'Rushing_First_Downs',
       'Percentage_of_Rushing_First_Downs', 'Rushing_More_Than_20_Yards',
       'Rushing_More_Than_40_Yards', 'Fumbles']].apply(pd.to_numeric, errors='coerce')
rush_stats_df.head()
```

```python
id = rec_stats['Player Id'].str.split('/', n=1, expand=True)
rec_stats['id'] = id[1]

name = rec_stats['Name'].str.split(', ', n=1, expand=True)
rec_stats['First Name'] = name[1]
rec_stats['Last Name'] = name[0]
```

```python
rec_stats_df = rec_stats[['id', 
                          'First Name', 
                          'Last Name',
                          'Position', 
                          'Year', 
                          'Team', 
                          'Games Played',
                           'Receptions', 
                          'Receiving Yards', 
                          'Yards Per Reception',
                           'Yards Per Game', 
                          'Longest Reception', 
                          'Receiving TDs',
                           'Receptions Longer than 20 Yards', 
                          'Receptions Longer than 40 Yards',
                           'First Down Receptions', 
                          'Fumbles']]
```

```python
rec_stats_df = rec_stats_df.rename(columns={'First Name':'First_Name', 
                          'Last Name':'Last_Name',
                          'Games Played':'Games_Played',
                          'Receiving Yards':'Receiving_Yards', 
                          'Yards Per Reception':'Yards_Per_Reception',
                           'Yards Per Game':'Yards_Per_Game', 
                          'Longest Reception':'Longest_Reception', 
                          'Receiving TDs':'Receiving_TDs',
                           'Receptions Longer than 20 Yards': 'Receptions_Longer_than_20_Yards', 
                          'Receptions Longer than 40 Yards': 'Receptions_Longer_than_40_Yards',
                           'First Down Receptions':'First_Down_Receptions'})
rec_stats_df.head()
```

```python
rec_stats_df[['Receptions', 'Receiving_Yards', 'Yards_Per_Reception',
        'Longest_Reception', 'Receiving_TDs',
       'Receptions_Longer_than_20_Yards', 'Receptions_Longer_than_40_Yards',
       'First_Down_Receptions', 'Fumbles']] = rec_stats_df[['Receptions', 'Receiving_Yards', 'Yards_Per_Reception',
        'Longest_Reception', 'Receiving_TDs',
       'Receptions_Longer_than_20_Yards', 'Receptions_Longer_than_40_Yards',
       'First_Down_Receptions', 'Fumbles']].apply(pd.to_numeric, errors='coerce')
```

```python
engine = create_engine('sqlite:///nfl_db.sqlite', echo=False)
```

```python
basic_stats_df.to_sql(name='basic_stats', con=engine, if_exists='append', index=False)
```

```python
def_stats_df.to_sql(name='def_stats', con=engine, if_exists='append', index=False)
```

```python
rec_stats_df.to_sql(name='rec_stats', con=engine, if_exists='append', index=False)
```

```python
pass_stats_df.to_sql(name='pass_stats', con=engine, if_exists='append', index=True)
```

```python
rush_stats_df.to_sql(name='rush_stats', con=engine, if_exists='append', index=True)
```
